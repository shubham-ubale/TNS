## SECTION 2: Data Cleaning and Manipulation (25 marks)

### Q2.1 (5 marks): Remove duplicates
**Grading:** 2m check, 2m remove, 1m display

```python
print("Original:", df.shape)
print("Duplicates:", df.duplicated().sum())

df = df.drop_duplicates()
print("After removing:", df.shape)
```

### Q2.2 (5 marks): Handle missing values
**Grading:** 2m identify, 2m numerical, 1m categorical

```python
# Numerical - median
num_cols = df.select_dtypes(include=[np.number]).columns
for col in num_cols:
    if df[col].isnull().sum() > 0:
        df[col].fillna(df[col].median(), inplace=True)

# Categorical - mode
cat_cols = df.select_dtypes(include=['object']).columns
for col in cat_cols:
    if df[col].isnull().sum() > 0:
        df[col].fillna(df[col].mode()[0], inplace=True)

print("Missing after cleaning:", df.isnull().sum().sum())
```

## SECTION 9: Statistical Analysis (15 marks)

### Q9.1 (5 marks): Statistics for price
**Grading:** 1m each stat

```python
print(f"Mean: ₹{df['discounted_price'].mean():.2f}")
print(f"Median: ₹{df['discounted_price'].median():.2f}")
print(f"Mode: ₹{df['discounted_price'].mode()[0]:.2f}")
print(f"Std Dev: ₹{df['discounted_price'].std():.2f}")
print(f"Variance: ₹{df['discounted_price'].var():.2f}")
```

### Q9.2 (5 marks): Percentiles and IQR
**Grading:** 3m percentiles, 2m IQR

```python
q1 = df['rating'].quantile(0.25)
q2 = df['rating'].quantile(0.50)
q3 = df['rating'].quantile(0.75)
iqr = q3 - q1

print(f"Q1 (25th): {q1:.2f}")
print(f"Q2 (50th/Median): {q2:.2f}")
print(f"Q3 (75th): {q3:.2f}")
print(f"IQR: {iqr:.2f}")
```

### Q9.3 (5 marks): Detect outliers
**Grading:** 2m bounds, 2m identify, 1m display

```python
q1 = df['discounted_price'].quantile(0.25)
q3 = df['discounted_price'].quantile(0.75)
iqr = q3 - q1

lower = q1 - 1.5 * iqr
upper = q3 + 1.5 * iqr

outliers = df[(df['discounted_price'] < lower) | (df['discounted_price'] > upper)]

print(f"Total outliers: {len(outliers)} ({len(outliers)/len(df)*100:.2f}%)")
print(f"Lower bound: ₹{lower:.2f}")
print(f"Upper bound: ₹{upper:.2f}")
print("\nSample outliers:")
outliers.nlargest(5, 'discounted_price')[['product_name', 'discounted_price']]
```

### Bonus Q2 (5 marks): Summary function
**Grading:** 3m function, 2m testing

```python
def category_summary(cat_name):
    cat_data = df[df['category'] == cat_name]
    
    summary = {
        'Products': len(cat_data),
        'Avg Rating': cat_data['rating'].mean(),
        'Avg Price': cat_data['discounted_price'].mean(),
        'Price Range': f"₹{cat_data['discounted_price'].min():.0f} - ₹{cat_data['discounted_price'].max():.0f}",
        'Avg Discount': cat_data['discount_percentage'].mean()
    }
    return pd.Series(summary)

# Test with 2 categories
cats = df['category'].value_counts().head(2).index
for cat in cats:
    print(f"\n{cat}:")
    print(category_summary(cat))
```

### Bonus Q5 (5 marks): Recommendation function
**Grading:** 2m filter, 2m sort, 1m test

```python
def recommend_products(max_budget, min_rating, category=None):
    # Filter by budget and rating
    filtered = df[(df['discounted_price'] <= max_budget) & (df['rating'] >= min_rating)]
    
    # Filter by category if specified
    if category:
        filtered = filtered[filtered['category'] == category]
    
    # Sort by rating (desc) then discount (desc)
    recommendations = filtered.sort_values(['rating', 'discount_percentage'], 
                                          ascending=[False, False]).head(5)
    
    return recommendations[['product_name', 'rating', 'discounted_price', 'discount_percentage']]

# Test
print("Recommendations: Budget ₹1000, Rating >= 4.0")
recommend_products(max_budget=1000, min_rating=4.0)
```
