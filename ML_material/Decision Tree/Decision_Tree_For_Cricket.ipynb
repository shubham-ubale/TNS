```python
#Seperating target variable and input variables
target = cricket['Play_Cricket_Flag']
cricket_2 = cricket.copy()
cricket_2 = cricket_2.drop('Play_Cricket_Flag', axis =1)
cricket_2.shape
```

```python
cricket_2.head()
```

```python
# Defining the attributes
X = cricket_2
X["Gender"]=X["Gender"].apply(lambda x: 1 if x=="Male" else 0)
X["Height"]=X["Height"].apply(lambda x: 1 if x=="More than 5.5 Feet" else 0)
X["Class"]=X["Class"].apply(lambda x: 1 if x=="Class X" else 0)
```

```python
target.head()
```

```python
y = target
```

**Building a Tree**

```python
# Defining the decision tree algorithm

dtree=DecisionTreeClassifier()
dtree.fit(X,y)

print('Decision Tree Classifer Created')
```

**Testing**

```python
# Visualising the graph without the use of graphviz
# filled=True filles the color to indicate the majority class
# "precision" represents the number of decimal places needed for "gini"
# rounded:  When set to 'True', draw node boxes with rounded corners
# node_ids: When set to True, show the ID number on each node.
# proportion: When set to True, change the display of ‘values’ and/or ‘samples’ to be proportions and percentages respectively.

plt.figure(figsize = (20,20))
dec_tree = plot_tree(decision_tree=dtree, feature_names = cricket_2.columns,
                     class_names =["Do not play cricket", "Play Cricket"] , filled = True , precision = 4,
                     rounded = True, node_ids=True, proportion=False)

plt.savefig("one.png")


```

```python
# Setting Max_depth to "2"

plt.figure(figsize = (20,20))
dec_tree = plot_tree(decision_tree=dtree, feature_names = cricket_2.columns,
                     class_names =["Do not play cricket", "Play Cricket"] , filled = True , precision = 4,
                     rounded = True, node_ids=True, proportion=False, max_depth=2)

plt.savefig("one.png")
```
