An activation function is something applied to the value coming into node, which then transforms it into the value stored in that node, or the node output

```python
def relu(input):
    output = max(input, 0) # if negative → 0, if positive → keep it
    return output
```

```python
#dot product
node_0_value = (input_data * weights['node_0']).sum()
node_0_value = relu(node_0_value)
node_1_value = (input_data * weights['node_1']).sum()
node_1_value = relu(node_1_value)
print(node_0_value)
print(node_1_value)
```

```python
model_output = (hidden_later_values * weights['output']).sum()
print(model_output)
```

```python
input_data = np.array([2,3])
weights = {'node_0_0': np.array([2,4]),
           'node_0_1': np.array([4,-5]),
           'node_1_0': np.array([-1,1]),
           'node_1_1': np.array([2,1]),
           'output': np.array([3,-7])}
```

```python
# ---- HIDDEN LAYER 0 ----

# node_0_0: (2*2) + (3*4) = 4 + 12 = 16 → relu(16) = 16
node_0_0_input = (input_data * weights['node_0_0']).sum()
node_0_0_output = relu(node_0_0_input)
print(node_0_0_output)   # 16

# node_0_1: (2*4) + (3*-5) = 8 + (-15) = -7 → relu(-7) = 0
node_0_1_input = (input_data * weights['node_0_1']).sum()
node_0_1_output = relu(node_0_1_input)
print(node_0_1_output)   # 0

# combine both node outputs into hidden layer 0
hidden_layer_0 = np.array([node_0_0_output, node_0_1_output])
print(hidden_layer_0)    # [16, 0]

# ---- HIDDEN LAYER 1 ----

# node_1_0: (16*-1) + (0*1) = -16 + 0 = -16 → relu(-16) = 0
node_1_0_input = (hidden_layer_0 * weights['node_1_0']).sum()
node_1_0_output = relu(node_1_0_input)
print(node_1_0_output)   # 0

# node_1_1: (16*2) + (0*1) = 32 + 0 = 32 → relu(32) = 32
node_1_1_input = (hidden_layer_0 * weights['node_1_1']).sum()
node_1_1_output = relu(node_1_1_input)
print(node_1_1_output)   # 32

# combine both node outputs into hidden layer 1
hidden_layer_1 = np.array([node_1_0_output, node_1_1_output])  # [0, 32]

# ---- OUTPUT LAYER ----

# output: (0*3) + (32*-7) = 0 + (-224) = -224
# no relu here — regression problem, negative output is valid
model_output = (hidden_layer_1 * weights['output']).sum()
print(model_output)   # -224
```

```python
#Multiple data points 

def predict_row(input_data,weights):
    node_0_0_input = (input_data * weights['node_0_0']).sum()
    node_0_0_output = relu(node_0_0_input)
    # print(node_0_0_output)   # 16

    # node_0_1: (2*4) + (3*-5) = 8 + (-15) = -7 → relu(-7) = 0
    node_0_1_input = (input_data * weights['node_0_1']).sum()
    node_0_1_output = relu(node_0_1_input)
    # print(node_0_1_output)   # 0

    # combine both node outputs into hidden layer 0
    hidden_layer_0 = np.array([node_0_0_output, node_0_1_output])
    # print(hidden_layer_0)    # [16, 0]

    # ---- HIDDEN LAYER 1 ----

    # node_1_0: (16*-1) + (0*1) = -16 + 0 = -16 → relu(-16) = 0
    node_1_0_input = (hidden_layer_0 * weights['node_1_0']).sum()
    node_1_0_output = relu(node_1_0_input)
    # print(node_1_0_output)   # 0

    # node_1_1: (16*2) + (0*1) = 32 + 0 = 32 → relu(32) = 32
    node_1_1_input = (hidden_layer_0 * weights['node_1_1']).sum()
    node_1_1_output = relu(node_1_1_input)
    # print(node_1_1_output)   # 32

    # combine both node outputs into hidden layer 1
    hidden_layer_1 = np.array([node_1_0_output, node_1_1_output])  # [0, 32]

    # ---- OUTPUT LAYER ----

    # output: (0*3) + (32*-7) = 0 + (-224) = -224
    # no relu here — regression problem, negative output is valid
    model_output = (hidden_layer_1 * weights['output']).sum()
    # print(model_output)   # -224
    return model_output
```

```python
results= []
input_data = [[0,3],[2,4],[5,6]]
for input_data_row in input_data:
    print(input_data_row)
    print(predict_row(input_data_row,weights))
    results.append(predict_row(input_data_row,weights))
```

```python
def predict_row(input_data,weights ):
    node_0_value = (input_data * weights['node_0']).sum()
    node_0_value = relu(node_0_value)
    node_1_value = (input_data * weights['node_1']).sum()
    node_1_value = relu(node_1_value)

    hidden_later_values = np.array([node_0_value,node_1_value])
    output = (hidden_later_values * weights['output']).sum()
    print(output)
    return output

input_data = np.array([0,3])

target_actual = 3


weights_0 = {'node_0': np.array([2,1]),
           'node_1': np.array([1,2]),
           'output': np.array([1,1])}

model_output_0 = predict_row(input_data, weights_0)

error_0 = model_output_0 - target_actual
print(error_0)
```

```python
# tweaking the weights to reduce error

weights_1 = {'node_0': np.array([2,1]),
           'node_1': np.array([1,2]),
           'output': np.array([-1,1])}

model_output_1 = predict_row(input_data, weights_1)

error_1 = model_output_1 - target_actual
print(error_0)
print(error_1)
```

```python
from sklearn.metrics import mean_squared_error

input_data = [[0,3],[1,4]]
target_actual = [3,7]

model_output_0 = []
model_output_1 = []

for row in input_data:
    model_output_0.append(predict_row(row,weights_0))

    model_output_1.append(predict_row(row, weights_1))

mse_0 = mean_squared_error(target_actual, model_output_0)
mse_1 = mean_squared_error(target_actual, model_output_1)

print("Mean squared error with weights_0: %f" %mse_0)
print("Mean squared error with weights_1: %f" %mse_1)
```
