- Tokenization #Breaking text into individual words or characters — "Hello World" → ["Hello", "World"]
- Sequencing  #Converting words into numbers so the model can process them — "Hello" → 3, "World" → 7
- Padding # Making all sequences the same length by adding zeros — [3,7] → [0,0,3,7]
- Stemming #Cutting words to their root form roughly — "running" → "run", "flies" → "fli"
- Lemmatization #Converting words to their proper dictionary base form — "running" → "run", "flies" → "fly"

## Tokenization

```python
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
sentence = ['We like machine learning']
tokenizer = Tokenizer()
tokenizer.fit_on_texts(sentence)
```

The tokenizer scans all your text, counts how many times each word appears, then assigns 1 to the most common word, 2 to the next, and so on.

## Padding

```python
sentences = ['We like machine learning',
             'we are learning tokenization',
             'we are learning sequencing',
             'we are learning the padding technique',
             'machine learning and deep learning are fun']

tokenizer = Tokenizer(oov_token = '#OOV')
tokenizer.fit_on_texts(sentences)

sequences = tokenizer.texts_to_sequences(sentences)
sequences
```

Model needs all inputs same size — zeros are just fillers, they mean nothing!

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_sequences = pad_sequences(sequences)
padded_sequences
```

```python
padded_sequences = pad_sequences(sequences, padding = 'pre')
padded_sequences
```

```python
padded_sequences = pad_sequences(sequences, padding = 'post')
padded_sequences
```

## Lemmatization

wordnet is a large English dictionary database that NLTK needs to perform Lemmatization — without downloading it, the lemmatizer won't know what the base form of any word is.

```python
import nltk
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
lemmatizer.lemmatize('breaks')
```

```python
lemmatizer.lemmatize('break', pos = 'v')
```
